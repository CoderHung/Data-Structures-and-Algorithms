```python
def getMedian(a, b):
    
    if len(a) > len(b):
        a, b = b, a

    m, n = len(a), len(b)
    total = m + n
    half = total // 2

    low, high = 0, m

    while low <= high:
        mid1 = (low + high) // 2
        mid2 = half - mid1

        l1 = float('-inf') if mid1 == 0 else a[mid1 - 1]
        r1 = float('inf') if mid1 == m else a[mid1]

        l2 = float('-inf') if mid2 == 0 else b[mid2 - 1]
        r2 = float('inf') if mid2 == n else b[mid2]

        if l1 <= r2 and l2 <= r1:
            if total % 2 == 0:
                return (max(l1, l2) + min(r1, r2)) / 2.0
            else:
                return min(r1, r2)

        elif l1 > r2:
            high = mid1 - 1
        else:
            low = mid1 + 1

    return 0  

```

```python
def run_tests():
    print("Test 1:", getMedian([1, 3], [2, 4]) == 2.5)
    print("Test 2:", getMedian([1, 2], [3, 4]) == 2.5)
    print("Test 3:", getMedian([0, 0], [0, 0]) == 0.0)
    print("Test 4:", getMedian([1, 2], [1, 2]) == 1.5)
    print("Test 5:", getMedian([1, 5], [2, 3]) == 2.5)
    print("Test 6:", getMedian([1, 2, 3], [4, 5]) == 3.0)
    print("Test 7:", getMedian([1, 2], [3, 4, 5]) == 3.0)
    print("Test 8:", getMedian([2, 3, 5], [1, 4]) == 3.0)
    print("Test 9:", getMedian([1, 6], [2, 3, 4]) == 3.0)
    print("Test 10:", getMedian([1], [2, 3, 4, 5, 6]) == 3.5)
    print("Test 11:", getMedian([1, 2, 3, 4, 5], [6]) == 3.5)
    print("Test 12:", getMedian([], [1]) == 1)
    print("Test 13:", getMedian([1], []) == 1)
    print("Test 14:", getMedian([], [1, 2]) == 1.5)

run_tests()
```
