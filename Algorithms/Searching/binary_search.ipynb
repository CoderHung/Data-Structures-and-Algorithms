```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1
```

```python
def test_binary_search():
    print("Test 1:", binary_search([1, 2, 3, 4, 5], 3) == 2)
    print("Test 2:", binary_search([1, 2, 3, 4, 5], 1) == 0)
    print("Test 3:", binary_search([1, 2, 3, 4, 5], 5) == 4)
    print("Test 4:", binary_search([1, 2, 3, 4, 5], 6) == -1)
    print("Test 5:", binary_search([], 1) == -1)
    print("Test 6:", binary_search([10], 10) == 0)
    print("Test 7:", binary_search([1, 3, 5, 7, 9], 7) == 3)
    print("Test 8:", binary_search([1, 3, 5, 7, 9], 2) == -1)
    print("Test 9:", binary_search([-10, -5, 0, 5, 10], -5) == 1)
    print("Test 10:", binary_search([1, 2, 2, 2, 3], 2) in [1, 2, 3])  # Any index of 2 is valid

test_binary_search()
```
