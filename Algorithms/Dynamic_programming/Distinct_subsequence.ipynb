```python
def numDistinct(s: str, t: str):
        n, m = len(s), len(t)
        dp = [[-1 for _ in range(m)]for _ in range(n)]

        def count_match(i, j):
            if j < 0:
                return 1 
            elif i < 0:
                return 0
            
            if dp[i][j] != -1:
                return dp[i][j]
            
            if s[i] == t[j]:
                dp[i][j] = count_match(i-1, j-1) + count_match(i-1, j)
            else:
                dp[i][j] = count_match(i-1, j)

            return dp[i][j]
        
        return count_match(n-1, m-1)
```

```python
numDistinct("babgbag", "bag")  
```

```python
def test_numDistinct():
    assert numDistinct("rabbbit", "rabbit") == 3
    assert numDistinct("babgbag", "bag") == 5
    assert numDistinct("abc", "abc") == 1
    assert numDistinct("abc", "ac") == 1
    assert numDistinct("abc", "abcd") == 0
    assert numDistinct("aaaaa", "aa") == 10
    assert numDistinct("", "") == 1
    assert numDistinct("a", "") == 1
    assert numDistinct("", "a") == 0
    print("All tests passed.")


test_numDistinct()
```
